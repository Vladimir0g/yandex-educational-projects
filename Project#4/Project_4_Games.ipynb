## Шаг 5. Проверка гипотиз <a id="step_5"></a>
Шаг 5. Проверьте гипотезы
5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Сформируем гипотезы:
Н0: Средние пользовательские рейтинги платформ Xbox One и PC равны
Н1: Средние пользовательские рейтинги платформ Xbox One и PC не равны

Сделаем срезы для обеих платформ и сравним статистику пользовательских рейтингов

```python
xb = actual_data.query('platform == "XOne" and user_score > 0')
print('Статистика пользовательских рейтингов по Xbox One:')
display(xb['user_score'].describe())
print('='*120)
pc = actual_data.query("platform == 'PC' and user_score > 0")
print('Статистика пользовательских рейтингов по PC:')
display(pc['user_score'].describe())
```

Статистики рейтингов пользователей обих плотформ очень похожи. Медианные значения равны, средние очень близки по значению
Не большие отличия в размерах и среднем отклонении. Критерий проверки у становим на уровне 5%.

```python
alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind( # проверка гипотезы о равенстве средних генеральных совокупностей
    xb['user_score'], 
    pc['user_score'], 
    equal_var=True, # выполняется стандартная независимая проверка двух выборок, которая предполагает равные дисперсии
    alternative='two-sided') # Альтернативная гипотеза H1: M(X) != M(Y).

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Сформулируем гипотезы:
Н0: Средние пользовательские рейтинги жанров Action и Sports равны
Н1: Средние пользовательские рейтинги жанров Action и Sports не равны
Сделаем срезы для обоих жанров и сравним статистику пользовательских рейтингов

```python
action = actual_data.query('genre == "Action" and user_score > 0')
print('Статистика пользовательских рейтингов по жанру Action:')
display(action['user_score'].describe())
print('='*120)
sports = actual_data.query("genre == 'Sports' and user_score > 0")
print('Статистика пользовательских рейтингов по жанру Sports:')
display(sports['user_score'].describe())
```

```python
alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind( # проверка гипотезы о равенстве средних генеральных совокупностей
    action['user_score'], 
    sports['user_score'], 
    equal_var=True, # выполняется стандартная независимая проверка двух выборок, которая предполагает равные дисперсии
    alternative='two-sided') # Альтернативная гипотеза H1: M(X) != M(Y).

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Как вы сформулировали нулевую и альтернативную гипотезы;
Какой критерий применили для проверки гипотез и почему.

Нулевая и альтернативная гипотезы
Гипотиза №1
Постановка вопроса: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?"
Формулировка гипотизы:
Нулевая - Н0: Средние пользовательские рейтинги платформ XOne и PC равны.
Альтернативная  - Н1: Средние пользовательские рейтинги платформ XOne и PC не равны, значит проверку надо провести двухстороннюю (alternative='two-sided')
Гипотиза №2
Постановка вопроса: "Средние пользовательские рейтинги жанров Action и Sports разные?". Действуем по аналогии с прошлой гипотизой. Инвертируем условие! Получим:
Нулевая - Н0: Средние пользовательские рейтинги жанров Action и Sports равны
Альтернативная - Н1: Средние пользовательские рейтинги жанров Action и Sports не равны, значит проверку надо провести двухстороннюю alternative='two-sided'
P-значение (англ. P-value) — это мера вероятности того, что наблюдаемая разница могла произойти по случайности. Чем ниже значение p-значения, тем больше статистическая значимость наблюдаемой разницы.
Для подтверждения или опровержения необходимо установить уровень значимости для того или иного сравнения процессов, обычно его выбирают из размерного ряда 1 - 10 %, в зависимости от изученности и требовательности к данным. Обычно его устанавливают в среднем значении 5% или 0,05.
В модуле stats библиотеки scipy уже предусмотрены функции для выполнения двухвыборочных t-проверок. Это функция stats.ttest_ind. Если аргумент equal_var равен True, то выполняется стандартная независимая проверка двух выборок.
